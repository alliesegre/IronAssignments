```python
df_dedupped
```

```python
X = df_dedupped.drop(['songs_names','songs_ids','artists_names'], axis=1)
```

```python
X_prep = StandardScaler().fit_transform(X)
```

```python
X_scaled = pd.DataFrame(X_prep, columns=X.columns)
X_scaled
```

```python
kmeans = KMeans(n_clusters=6, random_state=1234)
kmeans.fit(X_scaled)

# print the cluster centers
clusters = kmeans.cluster_centers_

print(clusters)
```

Choose cluster number vs inertia

```python
K = range(2,20)
inertias = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                random_state=1234,
                n_init=20,
                algorithm='elkan',
                max_iter=25)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    
plt.figure(figsize=(16,8))
plt.plot(K, inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title(f'Elbow Method showing the optimal k within {list(K)}')
```

Check Silhouette

```python
K = range(2,20)
silhouettes = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                random_state=1234,
                n_init=10,
                algorithm='elkan',
                max_iter=10)
    kmeans.fit(X_scaled)
    silhouettes.append(silhouette_score(X_scaled,
                                     kmeans.predict(X_scaled)
                                    )
                   )

plt.figure(figsize=(16,8))
plt.plot(K, silhouettes, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title(f'Elbow Method showing the optimal k within {list(K)}')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15,8))
for k in [2, 3, 4, 5]:
    '''
    Create KMeans instance for different number of clusters
    '''
    km = KMeans(n_clusters=k,
                random_state=1234)
    q, mod = divmod(k, 2)
    '''
    Create SilhouetteVisualizer instance with KMeans instance
    Fit the visualizer
    '''
    
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(X_scaled)
```

5 appears to be the best number for k

```python
kmeans = KMeans(n_clusters=5,
               random_state=1234)
kmeans.fit(X_scaled)

clusters = kmeans.predict(X_scaled)
```

Input and Recommendation

```python
def song_reccomendation(song):
    #input_i = input('Give me a song title: ')
    song = song.lower()
    try:
        # check if song is on billboard hot 100
        if song in list(hotsongs_clean["title"]):
            # if song is in hotsongs, suggest another hotsong
            random = hotsongs_clean[hotsongs_clean["title"]!='drivers license'].sample(1)
            print("You might also like " + random['title'].values + " by " + random['artist'].values)   
        else:
            # if song is not hot, then based on cluster
            track_feat = pd.DataFrame(sp.audio_features(sp.search(q='track:'+song, type='track')['tracks']['items'][0]['id']))
            temp = track_feat.drop(['type','id','uri','track_href','analysis_url','duration_ms','time_signature'], axis=1)
            s_scale = StandardScaler().fit_transform(temp)
            song_pred = kmeans.predict(s_scale)[0]
            cluster_song_pred = kmeans.predict(s_scale)[0]
            song_foru = df_dedupped[df_dedupped['cluster_id']==cluster_song_pred].sample(1)
            song_name = song_foru['songs_names']
            artist_name =song_foru['artists_names']
            print(f'You like {song}? \n Then you will like {song_name} by {artist_name}!')
    except:
        print('Something did not work, try another song please!')
        
   
```

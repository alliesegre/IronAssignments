```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
#from urllib.parse import urlparse
#from urlparse import urljoin

url = "https://www.billboard.com/charts/hot-100"
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
print(soup.prettify())
```

```python
len_mov = len(soup.select("span.chart-element__information"))
artist =[]
song = []

for i in range(len_mov):
     artist.append(soup.select("span.chart-element__information__artist")[i].get_text())
     song.append(soup.select("span.chart-element__information__song")[i].get_text())
```

```python
ready_to_clean_1= pd.DataFrame({'artist':artist,'song':song})
```

```python
ready_to_clean_1
```
